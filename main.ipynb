```python
import pandas as pd
df=pd.read_csv('TCGA_GBM_LGG_Mutations_all.csv',na_values="--")
df
```

```python

# Handling Missing Data
df.drop(columns='Case_ID',inplace=True)
df.drop(columns='Project',inplace=True)
print(df.isnull().sum())
```

```python
# Data Cleaning
df.drop_duplicates(inplace=True) 
df.dropna(axis=0,inplace=True)
df.isnull().sum()
```

```python
df["IDH1"].unique()
```

```python
import pandas as pd


def split_age(input_str):
    parts = input_str.split()
    years = int(parts[0]) if parts[0].isdigit() else 0
    days = int(parts[2]) if len(parts) > 2 and parts[2].isdigit() else 0
    return years, days

df[['Years', 'Days']] = df['Age_at_diagnosis'].apply(lambda x: pd.Series(split_age(x)))

df10=df.copy()
df.drop(columns="Age_at_diagnosis",inplace=True)
df
```

```python
# Data Encoding
# We have Two major Values in the Attributes which is Categorical.
# Therefore we can change it to numerical values as 0s and 1s
from sklearn.preprocessing import LabelEncoder
label_encoders={}
for i in df.columns:
    if df[i].dtype==object:
        label_encoders[i]=LabelEncoder()
        df[i]=label_encoders[i].fit_transform(df[i])
df
```

```python
from sklearn.model_selection import train_test_split
X=df2
y=df['Grade']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# 1. Perceptron Classifier
from sklearn.linear_model import Perceptron
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from scipy.stats import uniform, randint
from sklearn.pipeline import Pipeline

preprocessor = StandardScaler()

param_dist = {
    'classifier__max_iter': randint(100, 1000),     
    'classifier__eta0': uniform(0.01, 0.1),         
    'classifier__tol': uniform(1e-3, 1e-1),         
}
perceptron = Perceptron()

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', perceptron)
])
random_search = RandomizedSearchCV(estimator=pipeline, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy')
random_search.fit(X_train, y_train)
best_params = random_search.best_params_
best_pipeline = Pipeline([
    ('preprocessor', StandardScaler()),
    ('classifier', Perceptron(max_iter=best_params['classifier__max_iter'], 
                               eta0=best_params['classifier__eta0'], 
                               tol=best_params['classifier__tol']))
])

best_pipeline.fit(X_train, y_train)


y_pred = best_pipeline.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred)
print("Accuracy: %.3f"%accuracy1)

```

```python
# 2. Logistic Regression
from sklearn.linear_model import LogisticRegression


preprocessor1 = StandardScaler()


param_dist = {
    'classifier__C': uniform(0.001, 100),
    'classifier__penalty': ['l1', 'l2'],
    'classifier__solver': ['liblinear', 'saga'],
    'classifier__max_iter': randint(1000, 2000),
    'classifier__tol': uniform(1e-5, 1e-3),
    'classifier__class_weight': [None, 'balanced']
}


lr = LogisticRegression()

# Define the pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor1),
    ('classifier', lr)
])

random_search = RandomizedSearchCV(estimator=pipeline, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy')


random_search.fit(X_train, y_train)

best_Param1 = random_search.best_params_

best_p1 = Pipeline([
    ('preprocessor', StandardScaler()),
    ('classifier', LogisticRegression())
])

best_p1.set_params(**best_Param1)
best_p1.fit(X_train, y_train)

y_pred = best_p1.predict(X_test)
accuracy2 = accuracy_score(y_test, y_pred)
print("Accuracy of LogisticRegression: %.3f"%accuracy2)

```

```python
# 3. Decision Tree Classification
from sklearn.tree import DecisionTreeClassifier
param_dist = {
    'classifier__criterion': ['gini', 'entropy'],
    'classifier__max_depth': randint(1, 1000),  # Adjust the range based on your data and computational resources
    'classifier__min_samples_split': randint(2, 20),
    'classifier__min_samples_leaf': randint(1, 20),
    'classifier__max_features': ['sqrt', 'log2', None],
    'classifier__class_weight': [None, 'balanced']
}
DTC=DecisionTreeClassifier()
pipeline=Pipeline([('preprocessor',StandardScaler()),('classifier',DTC)])
random_search=RandomizedSearchCV(estimator=pipeline,param_distributions=param_dist,cv=5,n_iter=10,scoring='accuracy')
random_search.fit(X_train,y_train)
best_param2=random_search.best_params_
best_p2=Pipeline([('preprocessor',StandardScaler()),('classifier',DecisionTreeClassifier())])
best_p2.set_params(**best_param2)
best_p2.fit(X_train,y_train)

y_pred=best_p2.predict(X_test)
accuracy3 = accuracy_score(y_test, y_pred)
print("Accuracy Score of DecisionTreeClassifier: %.3f"%accuracy3)
```

```python
# 4. Implementing Multiple Regression:

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lir=LinearRegression()
lir.fit(X_train,y_train)
y_pred=lir.predict(X_test)
r2Score = r2_score(y_test, y_pred)
print("Multiple regression R2_score: %.3f"%r2Score)
```

```python
# Comparing All Algos Used:
'''I have used Four classification algorithms
1. Perceptron: Classification Algo
2. Logistic Regression: Classification Algo
3. Decision Tree Classifier: Classification Algo
4. Multiple Regression: Regression Algo'''
import matplotlib.pyplot as plt

# Accuracy scores of Logistic Regression, Perceptron, and Decision Tree
accuracy_scores = {'Logistic Regression':accuracy1 , 'Perceptron': accuracy2, 'Decision Tree': accuracy3}

# R2 score of Multiple Regression
r2_score = r2Score

# Plotting accuracy scores
plt.figure(figsize=(10, 6))
plt.bar(accuracy_scores.keys(), accuracy_scores.values(), color=['blue', 'orange', 'green'])
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Models')
plt.ylim(0, 1)  # Set y-axis limits to range from 0 to 1
plt.axhline(y=r2_score, color='red', linestyle="--", label='R2 Score (Multiple Regression)')
plt.legend()
plt.show()
```
